```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("box.jpg")
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
```

```python
gray_img
```

```python
corners = cv2.goodFeaturesToTrack(image=gray_img,maxCorners=10,qualityLevel=0.01,minDistance=10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(img)
```
