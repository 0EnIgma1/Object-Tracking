```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#canny detection
img = cv2.imread("box.jpg")
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img,cmap='gray')
```

```python
#Canny edge detection
edges = cv2.Canny(image = img,threshold1 = 65, threshold2 = 65)
plt.imshow(edges)
```

```python
#Canny edge detection
edges = cv2.Canny(image = img,threshold1 = 125, threshold2 = 125)
plt.imshow(edges)
```

```python
#Canny edge detection
edges = cv2.Canny(image = img,threshold1 = 255, threshold2 = 255)
plt.imshow(edges)
```

```python
#Canny edge detection
edges = cv2.Canny(image = img,threshold1 = 0, threshold2 = 0)
plt.imshow(edges)
```

```python
#blur function
blur_img = cv2.blur(img,ksize=(5,5))
```

```python
#Canny edge detection
edges = cv2.Canny(image = blur_img,threshold1 = 65, threshold2 = 65)
plt.imshow(edges)
```

```python
#Grid detection
new_img = cv2.imread("bigdot.jpg")
plt.imshow(new_img)
```

```python
found,corners = cv2.findCirclesGrid(new_img,(15,15),cv2.CALIB_CB_SYMMETRIC_GRID)
```
